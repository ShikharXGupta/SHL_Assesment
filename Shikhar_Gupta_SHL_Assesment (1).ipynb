```python
# Imports
import whisper
import language_tool_python
from google.colab import files
import IPython.display as ipd
```

```python
# Upload audio file
print("Upload an audio file (wav/mp3/m4a)...")
uploaded = files.upload()
filename = list(uploaded.keys())[0]
ipd.display(ipd.Audio(filename))
```

```python
# Transcribe using Whisper
print("\nTranscribing with Whisper...")
model_whisper = whisper.load_model("base")
result = model_whisper.transcribe(filename)
transcribed_text = result["text"]
print(f"\n📝 Transcribed Text:\n\"{transcribed_text.strip()}\"")
```

```python
# Grammar Checking (Rule-Based)
print("\nChecking grammar using rule-based tool...")
tool = language_tool_python.LanguageTool('en-US')
matches = tool.check(transcribed_text)
```

```python
# Compute score out of 10 based on grammar issues
num_errors = len(matches)
num_words = len(transcribed_text.split())
error_rate = num_errors / num_words if num_words > 0 else 1
score = max(0, 10 - (error_rate * 10))  # subtract points for error rate
score = round(score, 2)
```

```python
print(f"\n🧠 Grammar Score: {score} / 10")
print(f"🔍 Detected Grammar Issues: {num_errors}")
for match in matches[:5]:  # Show top 5 grammar suggestions
    print(f" - {match.message} (Suggestion: {match.replacements})")
```
